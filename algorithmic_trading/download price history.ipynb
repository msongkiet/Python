Download SET history price

```python
# import library
import pandas as pd
import numpy as np
from tqdm import tqdm
import yfinance as yf

import warnings
warnings.filterwarnings('ignore')

# import date library
from datetime import date
today = date.today().strftime("%m_%d_%y")

print('pandas v.',pd.__version__,
      '\numpy v.',np.__version__,
      '\nyfinace v.',yf.__version__)
```

```python
# read SET symbol
df_symbol = pd.read_csv('Symbol_SET.csv')
tickers = df_symbol[df_symbol['Symbol'].str.contains('-R.BK') == False].reset_index(drop=True)
tickers.head()
```

```python
# download history function
def download_history(ticker = 'ADVANC.BK',period='1y',interval='1d'):
    """
    ticker: equity symbol in yahoo finance
    periods: 1d,5d,1mo,3mo,6mo,1y,2y,5y,10y,ytd,max (time period)
    interval: 1m,2m,5m,15m,30m,60m,90m,1h,1d,5d,1wk,1mo,3mo (trading interval)
    fetch data by interval (including intraday if period < 60 days)
    """
    data = yf.download(tickers = ticker, period = period, interval = interval)
    data['Symbol'] = ticker

    return data
```

```python
download_history()
```

```python
# define period and interval 

# valid periods: 1d,5d,1mo,3mo,6mo,1y,2y,5y,10y,ytd,max
period = 'max'

# fetch data by interval including intraday if period < 60 days)
# valid intervals: 1m,2m,5m,15m,30m,60m,90m,1h,1d,5d,1wk,1mo,3mo
intervals = ['1d','1wk']

for interval in intervals:
    df = pd.DataFrame()
    for ticker in tqdm(tickers['Symbol'][:]):
        # print(ticker)
        df_temp = download_history(ticker,period,interval)
        df = pd.concat([df,df_temp])

    print(df.head())
    print(df.tail())

    # save df to parquet file
    df.to_parquet(f'SET_history_price_{interval}_{today}.gzip',compression='gzip')
```
