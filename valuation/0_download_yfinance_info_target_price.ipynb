```python
stockTicker = pd.read_excel("tickers.xlsx", header=0, sheet_name="All SET")
stockTicker = stockTicker["ticker"].values.tolist()
```

```python
columns = ["Symbol","Recommendation","currentPrice","targetLowPrice","targetMedianPrice","targetMeanPrice","targetHighPrice","Beta"]
df = pd.DataFrame(columns=columns)

for ticker in stockTicker:
    try:
        info = yf.Ticker(f'{ticker}.BK').info
        recommendation = info["recommendationKey"]
        currentPrice = info["currentPrice"]
        targetLowPrice = info["targetLowPrice"]
        targetMedianPrice = info["targetMedianPrice"]
        targetMeanPrice = info["targetMeanPrice"]
        targetHighPrice = info["targetHighPrice"]
        beta = info["beta"]
        
        df = df.append(pd.DataFrame([[ticker,recommendation,currentPrice,targetLowPrice,targetMedianPrice,targetMeanPrice,targetHighPrice,beta]],columns=columns))
        print(f'{ticker} - Completed')
    except Exception as e:
        print(f'{ticker} - {e}')
    
    
df
```

```python
from datetime import datetime
dt_string = datetime.now().strftime("%Y%m%d_%H%M%S")

# df.to_pickle("info_20211230.pkl")
# df.to_excel("recommendation_20201230.xlsx")

df.to_pickle(f'info_{dt_string}.pkl')
df.to_excel(f'recommendation_{dt_string}.xlsx')
```
